```python
%matplotlib inline
import numpy as np
import pandas as pd
%load_ext autoreload
%autoreload 2
```

```python
import subprocess
import re
import random

path_train = 'data/train'

# decide rows to skip
nrows = subprocess.check_output(['wc', '-l', path_train])
nrows = int(re.sub("[^0-9]", "", nrows.decode(encoding='UTF-8')))
to_skip = set(range(1, nrows)) - set(random.sample(range(1, nrows), 10000))

# read data
pre_read = pd.read_csv(path_train, nrows=1000)
data = pd.read_csv(path_train, dtype=pre_read.dtypes.to_dict(), skiprows=list(to_skip))
```

```python
from helpers import ColumnExtractor, DateTimeExtractor
from sklearn.pipeline import Pipeline, FeatureUnion
from sklearn.preprocessing import OneHotEncoder

categorical_features = ['C1', 'site_id', 'site_domain', 'site_category', 'app_id', 'app_domain',
                        'app_category', 'device_id', 'device_ip', 'device_model', 'device_type',
                        'device_conn_type', 'C14', 'C15', 'C16', 'C17', 'C18', 'C19', 'C20', 'C21']

numeric_features = ['banner_pos']
datetime_features = ['hour']

pipeline = \
    FeatureUnion([
        ('extract_cat', Pipeline([
            ('extract', ColumnExtractor(categorical_features)),
            ('one_hot', OneHotEncoder(sparse=True))
        ])),
        ('extract_num', ColumnExtractor(numeric_features, categorical=False)),
        ('extract_time', Pipeline([
            ('extract', ColumnExtractor(datetime_features, categorical=False)),
            ('combo', FeatureUnion([
                ('extract_hour', DateTimeExtractor(mode='hour', format='%y%m%d%H')),
                ('extract_weekday', DateTimeExtractor(mode='weekday', format='%y%m%d%H'))
            ]))
        ]))
    ])

X = pipeline.fit_transform(data)
```

```python
from sklearn.manifold import TSNE
from time import time

t0 = time()
res_tsne = TSNE().fit_transform(X)
t1 = time()
print("t-SNE: %.2g sec" % (t1 - t0))
```

```python
from matplotlib import pyplot as plt

plt.scatter(res_tsne[:, 0], res_tsne[:, 1], c=data.click.values, cmap=plt.cm.Spectral)
plt.show()
```
